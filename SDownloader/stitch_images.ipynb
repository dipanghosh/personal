```python
import os
root = "D:\\SDO_Data\\AIA_193\\"

def get_imagelist(root_dir):
    list_of_images = []
    for path, subdirs, files in os.walk(root_dir):
        for name in files:
            if name.endswith(".jpg"):
                list_of_images.append (os.path.join(path, name)) 
    return list_of_images
```
