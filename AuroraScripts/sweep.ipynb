```python
from datetime import date, timedelta, datetime
import pandas as pd
from tqdm import tqdm_notebook
import os

def get_days(start, end):
    d1 = datetime.strptime(str(start), "%Y%m%d")
    d2 = datetime.strptime(str(end), "%Y%m%d")
    delta = d2 - d1         # timedelta
    return delta.days
    
def dateseq_gen(start, end):
    d1 = datetime.strptime(str(start), "%Y%m%d")
    d2 = datetime.strptime(str(end), "%Y%m%d")

    delta = d2 - d1         # timedelta

    for i in range(delta.days + 1):
        yield ((d1 + timedelta(i)).strftime("%Y%m%d"))
```

```python
def filter_garbage(e):
    if e in ["","\n","[","]", " "]:
        return False
    else:
        return True
    
pred_from_log = {}
logfile = open(logfilepath, "r")
for line in logfile.readlines():
    if line.rstrip():
        line = "".join((filter(filter_garbage, line)))
        date = int(line.split(",")[0])
        pred = line.split(",")[1:]
        if pred[0] == "":
            pred_from_log[date] = []
        else:
            pred = map(float, pred)
            pred_from_log[date] = pred
            
df = pd.DataFrame.from_dict(pred_from_log, orient='index')
df.to_csv("2017.csv")
```
